```python
from bandit import *
import plotly.express as px
```

```python
list1 = run_kbandit(2000,10,1,1,10)
running_mean = []
for i in range(0,len(list1[1])):
    curr_list = list1[1][0:i]
    if len(curr_list) > 0: running_mean.append(sum(curr_list)/len(curr_list))

running_mean
```

```python
fig = px.scatter(x=list(range(0,len(running_mean))), y=running_mean, labels={'x': 'i', 'y': 'Value'})

# Show the plot
fig.show()
```
